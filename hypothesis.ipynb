# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here:

from scipy.stats import ttest_1samp

# Considering hourly rate only from hourly_df.
hourly_wage = hourly_df["Hourly Rate"]

# we select a value for alpha of 0.05 (p-value threshold, significance level)
# Two sided test: 
# Null hypothesis = The hourly wage is $30/hr.
# Alternative hypothesis = The hourly wage <> $30/hr.

stat, pval = ttest_1samp(hourly_wage, 30)

print("stat ", stat)
print("pval ", pval)
```

```python
# CONCLUSION: The p value is close to 0 which is less than our alpha(0.05). 
# Hence we can reject Null hypothesis. It means the hourly wage is different than $30/hr. 
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our Null hypothesis is less than or equal to, then to reject, the test statistic must be positive (rejects the <) and the p-value below 0.05 (rejects the =).

```python
# Your code here:

# One sided hypothesis
# Null Hypothesis: Salaries of police department > 86000.
# Alternative hypothesis: The salaries are < 86000.

police_salary = salaries[(salaries["Department"] == "POLICE") & (salaries["Salary or Hourly"] == "Salary")]
display(police_salary.head())

stat2, pval2 = ttest_1samp(police_salary["Annual Salary"], 86000)

print("stat ", stat2)
print("pval ", pval2/2)
```

```python
# CONCLUSION: The p value is close to 0 which is less than our alpha(0.05). 
# Hence we can reject Null hypothesis. It means the annual salary of police department is less than 86000.
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:

# Goal: Compute 95% confidence level for mean hourly wage of all hourly workers.

import scipy.stats

hourly_wage = hourly_df["Hourly Rate"]

confidence_level = 0.95

degrees_freedom = len(hourly_wage) - 1
print(degrees_freedom)

sample_mean = np.mean(hourly_wage)
print(sample_mean)

sample_standard_error = scipy.stats.sem(hourly_wage)
print(sample_standard_error)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)

print(confidence_interval)
```

```python
# CONCLUSION: As per above results, the confidence interval of provided sample lies between 34.58 and 34.87.
```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
# Your code here:
police_salaries = police_salary["Annual Salary"]

degrees_freedom2 = len(police_salaries) - 1
print(degrees_freedom2)

sample_mean2 = np.mean(police_salaries)
print(sample_mean2)

sample_standard_error2 = scipy.stats.sem(police_salaries)
print(sample_standard_error2)


confidence_interval2 = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom2, 
                                             sample_mean2, 
                                             sample_standard_error2)

print(confidence_interval2)
```

```python
# CONCLUSION: As per above results, the confidence interval of provided sample lies between 88534.11 and 89134.11.
```
